# Statistical Data Analysis:-

```python
num=df.select_dtypes(include='number')
print("The mean values of the data are:\n ",num.mean())
print("**********************************************")
print("The median values of the data are:\n ",num.median())
print("**********************************************")
print("The mode values of the data are:\n ",num.mode().iloc[0])
print("**********************************************")
print("The range of the data is:\n ",num.max()-num.min())
print("**********************************************")
print("The IQR of the data is:\n ",num.quantile(0.75)-num.quantile(0.25))
print("**********************************************")
print("The standard deviations of the data are:\n ",num.std())
```

# Data Preprocessing:

```python
num=df.select_dtypes(include='number')

#Method1: Rescaling using MinMaxScaler
print('Min Max Scaled data::')
df1=preprocessing.MinMaxScaler(feature_range=(0,1))
print(df1.fit_transform(num).round(2))
print('\n=============================')

#Method2: Normalizing
print('L1 Normalized Data::')
df2=preprocessing.normalize(num,norm='l1')
print(df2.round(2))
print("\n=============================")

#Methhod3: Binarizing
print("Binarizing data::")
df3=preprocessing.Binarizer(threshold=5).transform(num)
print(df3)
print("\n==============================================================")

#method4: Standardizing
print('Standardizing data::')
df4=preprocessing.StandardScaler().fit(num)
print(df4.transform(num))
```

```python
#Method6: Label Encounter on Category
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
df5=labelencoder.fit_transform(df['Category'])
df5
```

# Data Visualization:-

### 1. Line Plot (Sales & Profit)

```python
plt.plot(df['Sales'],df['Profit'],color='goldenrod',marker='*',linestyle=':')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

### 2. Multiple SubPlots in Separate Figure of different columns

```python
fig,axes=plt.subplots(2,2,figsize=(11,8))

axes[0,0].plot(df['Sales'],'r')
axes[0,0].set_title('Sales')

axes[0,1].plot(df['PostalCode'],'b')
axes[0,1].set_title('Postal Code')

axes[1,0].plot(df['Profit'],'g')
axes[1,0].set_title('Profit')

axes[1,1].plot(df['Quantity'],'y')
axes[1,1].set_title('Quantity')

fig.suptitle('Numeric Columns')
plt.show()
```

### 3. Axes Object on (RowID X Profit & Sales)

```python
fig=plt.figure()
ax=fig.add_axes([0,0,2,1])
l1=ax.plot(df['Sales'],df['RowID'],'ys--')
l2=ax.plot(df['Profit'],df['RowID'],'go--')
ax.legend(labels=('Sales','Profit'),loc='upper right')
ax.set_xlabel('Sales & Profit')
ax.set_ylabel('RowId')
plt.show()
```

### 4. Bar Graph (Category & Sales)

```python
dfn=df.tail(30)
category=dfn['Category']
sales=dfn['Sales']
plt.bar(category,sales,color='mediumpurple')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Bar Graph')
plt.show()
```

### 5. Pie Chart (Sales & City)

```python
dfm=df.tail(10)
explode=[0.2,0.3,0.2,0.1,0.2,0.3,0.1,0.2,0.3,0.1]
plt.pie(dfm['Sales'],labels=dfm['City'],autopct='%1.1f%%',explode=explode)
plt.title('Pie Chart')
plt.show()
```

### 6. Donut Chart (Sales & City)

```python
plt.pie(dfm['Sales'],labels=dfm['City'],autopct='%1.1f%%',shadow=True)
circle=plt.Circle((0,0),0.75,color='black',fc='white',linewidth=1.25)
Figure=plt.gcf()
Figure.gca().add_artist(circle)
plt.title('Donut Chart')
plt.show()
```

### 7. Scatter Plot (Postal Code & RowID)

```python
dfp=df.sample(50)
plt.scatter(dfp['PostalCode'],dfp['RowID'],s=160,c='darkviolet',marker='*')
plt.xlabel('Postal Code')
plt.ylabel('RowID')
plt.title('Scattter PLot')
plt.show()
```

### 8. Area Chart (RowId & Sales)

```python
dff=df.head(10)
plt.fill_between(dff['RowID'],dff['Sales'],color='lightpink')
plt.plot(dff['RowID'],dff['Sales'],color='black')
plt.title('Area Plot')
plt.xlabel('RowId')
plt.ylabel('Sales')
plt.show()
```

### 9. Histogram (Segment)

```python
df['Segment']=df['Segment'].astype(str)
plt.hist(df['Segment'],facecolor='aqua',edgecolor="g",bins=3)
plt.xlabel("Segment Bins")
plt.ylabel('Frequency')
plt.title("Histogram")
plt.show()
```

### 10. Multiple Histogram( Category & ShipMode)

```python
df['Category']=df['Category'].astype(str)
df['ShipMode']=df['ShipMode'].astype(str)
plt.hist(df['Category'],facecolor='purple',alpha=0.3,edgecolor='green',bins=3,label='Category')
plt.hist(df['ShipMode'],facecolor='green',alpha=0.3,edgecolor='purple',bins=4,label='Ship Mode')
plt.xticks(rotation=45,ha='right')
plt.title('Multiple Histogram')
plt.legend()
plt.show()
```

### 11. Box Plot ( Postal Code)

### 12. Box Plot using Outliers (Not from Dataset)

### 13. Multivariate Analysis Using Box Plot

```python
df['Category']=df['Category'].astype(str)
grouped=df.groupby('Category')
fig,axes=plt.subplots(1,2,figsize=(12,4))

#Row ID
df.boxplot(column='RowID',by="Category",ax=axes[0])
axes[0].set_title("RowID")

#Postal Code
df.boxplot(column='PostalCode',by="Category",ax=axes[1])
axes[1].set_title("Postal Code")

plt.tight_layout()
plt.show()
```
